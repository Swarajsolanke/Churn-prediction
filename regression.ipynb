```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder,OneHotEncoder
import pickle
```

```python
data=pd.read_csv("D:\krishnaik course projects\Churn prediction\Churn_Modelling.csv")
data.head()
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
label_gender_encoder=LabelEncoder()
data['Gender']=label_gender_encoder.fit_transform(data['Gender'])
data
```

```python
# geography one hot encoder 
one_hot_encoder=OneHotEncoder(handle_unknown='ignore')
geo_data=one_hot_encoder.fit_transform(data[['Geography']]).toarray()
geo_data_df=pd.DataFrame(geo_data, columns=one_hot_encoder.get_feature_names_out(['Geography']))
geo_data_df
```

```python
#concatinate the features 
geo_data_df=pd.concat([data.drop('Geography',axis=1),geo_data_df],axis=1)
```

```python
geo_data_df
```

```python
X=geo_data_df.drop('EstimatedSalary',axis=1)
y=geo_data_df['EstimatedSalary']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=34)
```

```python
scaler1=StandardScaler()
X_train_scaled=scaler1.fit_transform(X_train)
y_trained_scaled=scaler1.transform(X_test)
X_train_scaled
```

```python
# pickle file 
with open('one_hot_encoder1.pkl','wb') as file:
    pickle.dump(one_hot_encoder,file)
with open('label_encoder_gender1.pkl','wb') as file:
    pickle.dump(label_gender_encoder,file)
with open('scaler1.pkl','wb') as file:
    pickle.dump(scaler1,file)
```

```python
# ANN with regression 
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# build model 
model=Sequential(
    [
        Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
        Dense(32,activation='relu'),
        Dense(1) # output layers for regression 

    ]
)
#comiple the model 
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
# set up the log 
log_dir='regreesion/fit/'+ datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorboard_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
earlystopping=EarlyStopping(monitor='val_loss',patience=8,restore_best_weights=True)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,callbacks=[earlystopping,tensorboard_callbacks])
```

```python
# evaluate the model 
test_loss,test_mae=model.evaluate(X_test,y_test)
print(f'test MAE: {test_mae}')
```
